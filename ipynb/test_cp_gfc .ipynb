# 反了吧 应该流入是N1 流出是N2吧
- Let `inbound edges` as $N_2$, `out` as $N_1$,
- net inventory as $d$,
we see, 

\begin{align*}
\begin{aligned}
& \sum_{j \in N_1} x_j-\sum_{j \in N_2} x_j \leq d, \\
& l_j y_j \leq x_j \leq u_j y_j, \quad y_j \in\{0,1\}, \quad j \in N_1 \cup N_2
\end{aligned}
\end{align*}

where each $x_j$ is the total flow summarizing all SKUs
